```
mymeshrdd1= mymeshrdd.map(lambda x: (x[1].split('.')))\
.map(lambda x: (x[0], int(1)))\
.reduceByKey(lambda x,y: x+y)
```

```
mymeshrdd1.collect()
```

```
mymeshDF= spark.createDataFrame (mymeshrdd1)
```

```
mymeshDF.printSchema()
```

QUESTION 6

```
CompletedstudyDF=clinicaltrial_2021.filter(clinicaltrial_2021.Completion.endswith("2021")).filter(clinicaltrial_2021.Status=="Completed").groupBy("Completion").count().sort(desc("count"))
```

```
CompletedstudyDF2=CompletedstudyDF.orderBy("Completion").withColumn("Completion", split(col("Completion"), " ").getItem(0))
```

```
CompletedstudyDF2.sort("Completion", inplace=True).show()
```

```
from pyspark.sql.functions import unix_timestamp
Completed_Studies_by_month=CompletedstudyDF2.sort(unix_timestamp(col("Completion"), "MMM"))
Completed_Studies_by_month.show()
```
